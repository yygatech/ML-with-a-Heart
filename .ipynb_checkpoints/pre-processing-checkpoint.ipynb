```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder

train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
train[targetCol] = np.log10(train[targetCol])
allData = pd.concat([train, test], axis =0, join ='outer', ignore_index =False)
allData.drop("Id", axis = 1, inplace = True)
# test.drop("Id", axis = 1, inplace = True)

idCol = 'Id'
targetCol = 'SalePrice'
```

```python
allData.head(5)
# print(train.columns)
```

```python
naDict = {}
for col in train:
    naDict[col] = train[col].isnull().sum() / len(train[col])

naDF = pd.DataFrame({'missing ratio':naDict}).sort_values('missing ratio', ascending=False)
naDF.head(20)
```

```python
f, ax = plt.subplots(figsize=(15, 12))
plt.xticks(rotation='90')
sns.barplot(x=naDF.index, y='missing ratio', data = naDF)
plt.xlabel('Features', fontsize=15)
plt.ylabel('Percent of missing values', fontsize=15)
plt.title('Percent missing data by feature', fontsize=15)
plt.show()
```

```python
mostMissing = ['PoolQC', 'MiscFeature', 'Alley', 'Fence', 'FireplaceQu', 'LotFrontage']
# train['Street'].corr(train['SalePrice'])
```

```python
trainNum = train.shape[0]
testNum = test.shape[0]

oeDF = allData.copy()

for column in allData:  
    if column == idCol or column == targetCol:
        continue
    
    if allData[column].dtype == 'O':
        temp = pd.get_dummies(oeDF[column], prefix=column, dummy_na=True)
        for c in temp:
            temp = temp.rename(columns = {c: c.replace('.', 'd')})
            
        oeDF = pd.concat([oeDF, temp], axis=1)
        oeDF.drop([column], axis=1, inplace=True)

    else:
        temp = train.ix[:trainNum-1, column].fillna(train[column].mean())
        mean = temp.mean()
        std = temp.std()
        
        oeDF[column] = oeDF[column].fillna(mean)
        oeDF[column] = (oeDF[column] - mean) / std
        
indexedTrain = oeDF.head(train.shape[0])
indexedTest = oeDF.tail(test.shape[0])
```

```python
corrmat = indexedTrain.corr()
plt.subplots(figsize=(15,12))
sns.heatmap(corrmat, vmax=0.9, square=True)
plt.show()

# temp = corrmat.sort_values(targetCol, ascending=False)

for c in mostMissing:
    if c != 'LotFrontage':
        print('correlation between {} and {} is {}.'.format(c, targetCol, corrmat.loc[c+'_nan', targetCol]))
    else:
        print('correlation between {} and {} is {}.'.format(c, targetCol, corrmat.loc[c, targetCol]))
        
# temp = abs(corrmat[targetCol])
# # temp = temp.sort_values(targetCol, ascending=False)
# # selectedFeatures = temp.index.head(20)

# k = 20
# selectedFeatures = temp.nlargest(k, targetCol)[targetCol].index


k = 50
temp = abs(corrmat)
temp = temp.sort_values(targetCol, ascending=False)
selectedFeatures = list(temp.index)[:k]
# print(selectedFeatures, len(selectedFeatures))



            
# temp
# for col in indexedTrain:
#     corr = indexedTrain[col].corr(indexedTrain['SalePrice'])
#     print('correlation between {} and {} is {}.'.format(col, 'SalePrice',corr))

# corrmat[targetCol]
# f, ax = plt.subplots(figsize=(15, 12))
# plt.xticks(rotation='90')
# sns.barplot(x = corrmat.index, y=corrmat[targetCol])
# plt.xlabel('Features', fontsize=15)
# plt.ylabel('Correlation with SalePrice', fontsize=15)
# plt.title('Correlation with SalePrice by feature', fontsize=15)
# plt.show()
```

```python
oeDF.head(train.shape[0]).to_csv('newOETrainAll.csv')
oeDF.tail(test.shape[0]).to_csv('newOETestAll.csv')

oeDF.head(train.shape[0])[selectedFeatures].to_csv('newOETrainTop{}.csv'.format(k))
oeDF.tail(test.shape[0])[selectedFeatures].to_csv('newOETestTop{}.csv'.format(k))
```

```python
trainNum = train.shape[0]
testNum = test.shape[0]

leDF = allData.copy()

for column in leDF:  
    if column == idCol or column == targetCol:
        continue
    
    if leDF[column].dtype == 'O':
        if leDF[column].isnull().values.any():
            leDF[column] = leDF[column].fillna(value = ' ')
        
        le = LabelEncoder()
        leDF[column] = le.fit_transform(leDF[column])
        
    else:
        temp = train.ix[:trainNum-1, column].fillna(train[column].mean())
        mean = temp.mean()
        std = temp.std()
        
        leDF[column] = leDF[column].fillna(mean)
        leDF[column] = (leDF[column] - mean) / std
        
indexedTrain = leDF.head(train.shape[0])
indexedTest = leDF.tail(test.shape[0])
```

```python
corrmat = indexedTrain.corr()
plt.subplots(figsize=(15,12))
sns.heatmap(corrmat, vmax=0.9, square=True)
plt.show()

# for c in mostMissing:
#     if c != 'LotFrontage':
#         print('correlation between {} and {} is {}.'.format(c, targetCol, corrmat.loc[c+'_nan', targetCol]))
#     else:
#         print('correlation between {} and {} is {}.'.format(c, targetCol, corrmat.loc[c, targetCol]))
        
# temp = abs(corrmat[targetCol])
# # temp = temp.sort_values(targetCol, ascending=False)
# # selectedFeatures = temp.index.head(20)

# k = 20
# selectedFeatures = temp.nlargest(k, targetCol)[targetCol].index


k = 20
temp = abs(corrmat)
temp = temp.sort_values(targetCol, ascending=False)
selectedFeatures = list(temp.index)[:k]
# print(selectedFeatures, len(selectedFeatures))
```
